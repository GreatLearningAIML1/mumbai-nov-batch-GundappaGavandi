```python
laliga = pd.read_csv('Laliga.csv', skiprows=0, index_col = "Pos") # Skip the first 1 rows of the file
laliga.replace({'-':0}, inplace = True)
laliga.head(5)
```

#Print all the teams which have started playing between 1930-1980.

```python
laliga['SinceYear'] = laliga['Debut'].astype(str).str[:4]
laliga.SinceYear = pd.to_numeric(laliga.SinceYear)
laliga_GT1930 = laliga[1930 <=(laliga["SinceYear"] )]
laliga_GT1930_LT1980 = laliga_GT1930[(laliga_GT1930["SinceYear"] <= 1980 )]
laliga_GT1930_LT1980.sort_values(['SinceYear'], ascending = True, inplace = True)
```

```python
laliga_GT1930_LT1980[["Team","SinceYear"]]
```

```python
#Print the list of teams which came Top 5 in terms of points
laliga.Points = pd.to_numeric(laliga.Points)
laliga.sort_values(["Points"], ascending = False, inplace = True)
laliga[['Team','Points']].head(5)
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.
#Using the same function, find the team which has maximum and minimum goal difference.
#Goal_diff_count = GoalsFor - GoalsAgainst

laliga.GoalsFor = pd.to_numeric(laliga.GoalsFor)
laliga.GoalsAgainst = pd.to_numeric(laliga.GoalsAgainst)

def Goal_diff_count():
    laliga['GoalDifference']=laliga['GoalsFor'] - laliga['GoalsAgainst']
       
    return laliga[['Team','GoalDifference']], laliga[laliga['GoalDifference'] == laliga['GoalDifference'].max()], laliga[laliga['GoalDifference'] == laliga['GoalDifference'].min()]
(teams_and_goal_diff, team_max_goal, team_min_goal)=Goal_diff_count()

teams_and_goal_diff
```

```python
#Create a new column with name “Winning Percent” and append it to the data set
#Percentage of Winning = (GamesWon / GamesPlayed)*100
laliga["Winning Percent"] = (pd.to_numeric(laliga.GamesWon)/pd.to_numeric(laliga.GamesPlayed))*100

#If there are any numerical error, replace it with 0%
laliga["Winning Percent"].fillna(0,inplace = True)

#Print the top 5 teams which has the highest Winning percentage
laliga.sort_values(["Winning Percent"], ascending = False, inplace = True)
laliga[['Team','Winning Percent']].head(5)
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions
laliga['Points'] = pd.to_numeric(laliga['Points'])
gp_pos = laliga.groupby('BestPosition').sum()
gp_pos[['Points']]
```
